```python
import random
import numpy as np
```

```python
def fakeData(op,max_val, min_val, close_val):
    minf = False
    maxf = False
    lst = []
    lst.append(op)
    actv = op
    for i in range(1, 28799):
        if i < 10000:
            actv = addRandNum(actv,max_val,min_val)
            lst.append(actv)
        elif i < 15000 and i >=10000:
            rnd_flag = round(random.uniform(0,1000))
            if rnd_flag > 250 and rnd_flag < 750 and not(minf):
                lst,minf,it = triggerMin(lst,actv,min_val,i)
                actv = min_val
                i =  it
            elif rnd_flag > 750 and not(maxf):
                lst,maxf,it = triggerMax(lst,actv,max_val,i)
                actv = max_val
                i =  it
            else:
                actv = addRandNum(actv,max_val,min_val)
                lst.append(actv)
        elif i < 20000 and i >= 15000:
            if not(maxf):
                lst,maxf,it = triggerMax(lst,actv,max_val,i)
                actv = max_val
                i =  it
            else:
                actv = addRandNum(actv,max_val,min_val)
                lst.append(actv)
                
        elif i >= 20000 and i < 25000:
            if not(minf):
                lst,minf,it = triggerMin(lst,actv,min_val,i)
                actv = min_val
                i = it
            else:
                actv = addRandNum(actv,max_val,min_val)
                lst.append(actv)
        else:
            if actv > close_val:
                miss = actv-close_val
                amt = 28799 - len(lst)
                if amt != 0:
                    add = miss / amt
                    for j in range (1,amt):
                        actv = actv - add
                        lst.append(round(actv,2))
            else:
                miss = close_val - actv
                amt = 28799 - len(lst)
                if amt != 0:
                    add = miss / amt
                    for j in range (1,amt):
                        actv = actv + add
                        lst.append(round(actv,2))
            i = 28800
    lst.append(close_val)
    return lst
            
            
                
            
            
def addRandNum(actv,max_val,min_val):
    rnd = round(random.uniform(0,0.01),3)
    vrc = rnd if random.uniform(0, 0.01) > 0.05 else rnd * -1
    if actv + vrc > max_val:
        actv = max_val
    elif actv + vrc < min_val:
        actv = min_val
    else:
        actv = actv + vrc
    return round(actv,2)

def triggerMin(lstv, actv, minv, it):
    while actv > minv:
        it +=1
        rnd = round(random.uniform(0,0.01),3)
        actv = actv - rnd if actv - rnd > minv else minv
        lstv.append(round(actv,2))
    return lstv, True, it

def triggerMax(lstv, actv, maxv,it):
    while actv < maxv:
        it+=1
        rnd = round(random.uniform(0,0.01),3)
        actv = actv + rnd if actv + rnd < maxv else maxv
        lstv.append(round(actv,2))
    return lstv, True, it
```

```python
f =  open("/home/tbeppe/Documents/tcc/pre-processing/dayone.csv","r")
```

```python
arr_ticks = []
```

```python
for l in f.readlines():
    tick_dict = {}
    splt = l.split(",")
    tick_dict["tick"] = splt[1]
    tick_dict["Name"] = splt[2]
    tick_dict["Op_value"] = splt[4]
    tick_dict["Max_value"] = splt[5]
    tick_dict["Min_value"] = splt[6]
    tick_dict["Close_value"] = splt[7]
    arr_ticks.append(tick_dict)
f.close()
```

```python
for c in arr_ticks:
    c["Day_values"] = fakeData(float(c["Op_value"]),float(c["Max_value"]),float(c["Min_value"]),float(c["Close_value"]))
    
```
